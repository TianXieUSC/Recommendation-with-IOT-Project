```python
del feature['Unnamed: 0']
feature.head()
```

```python
for i in range(31):
    
    print('**********',feature_hash[i],'**********')
    fig,ax=plt.subplots()
    fig.set_size_inches(10,6)
    n,bins,patches=plt.hist(feature['genre%d'%i].values,100)
    plt.title('Feature%d'%i)
    plt.grid(True)
    plt.show()
    print(bins)
    print(n)
    print('\n')
    plt.close()
```
